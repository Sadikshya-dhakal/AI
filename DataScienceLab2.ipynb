In this experiment, a set of 1000 random data points was generated using the NumPy normal() function, which produces values following a normal distribution with mean 0 and standard deviation 1. The generated data was then stored in a pandas DataFrame to allow structured handling and analysis. Using the df.describe() function, key statistical measures such as mean, standard deviation, minimum, maximum, and quartiles were computed.
```python
# Generating Sample Data

data = np.random.normal(loc=0, scale=1,size=1000)
df=pd.DataFrame(data)
df.describe()
```

# Plotting Histogram
This section of the code is used to visualize the distribution of the generated data by plotting a histogram. The plt.hist() function divides the dataset into 30 bins and displays the relative frequency of the values, with density=True ensuring the histogram is normalized. The transparency level of the bars is adjusted using alpha=0.7, and the color is set to blue for clarity. The title and axis labels are added to clearly describe the plot and its components. Finally, plt.show() renders the histogram, allowing for visual interpretation of the dataâ€™s distribution pattern.

```python
# Plottig Histogram

plt.hist(data, bins=30, density=True, alpha=0.7, color='blue')
plt.title('Histogram of Sample Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

# Empirical CDF
This portion of the code is used to plot the empirical cumulative distribution function (ECDF) of the sample data. The plt.hist() function is applied with the parameters density=True and cumulative=True to convert the histogram into a cumulative distribution, showing the proportion of data points less than or equal to each value. The plot uses 30 bins for smooth visualization, and alpha=0.7 provides slight transparency, while the bars are colored green. The title and axis labels clearly indicate that the graph represents cumulative probability against data values. Finally, plt.show() displays the ECDF, allowing for a clear visual interpretation of how the data accumulates across its range.

```python
# Plottig empirical CDF

plt.hist(data, bins=30, density=True,cumulative=True, alpha=0.7, color='green')
plt.title('Empirical CDF of Sample Data')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.show()
```

# Normal Distribution
In this section, a normal distribution is simulated using NumPy. A dataset of 100,000 values is generated with a mean of 100 and a standard deviation of 5. A histogram of these values is plotted using Matplotlib to visualize the distribution. Additionally, a red dashed line is drawn at the mean value to indicate the central tendency of the data. This plot helps to understand the shape, spread, and symmetry of a normal distribution.

```python
# Normal Distribution
Mean=100
Standard_deviation=5
size=100000
values=np.random.normal(Mean, Standard_deviation, size)
plt.hist(values, 100)
plt.axvline(values.mean(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

# Exponential Distribution
In this section, an exponential distribution is simulated using NumPy. A dataset of 1000 values is generated with a scale parameter of 2. The distribution is visualized using Seabornâ€™s histplot(), which plots a histogram along with a kernel density estimate (KDE) to show the dataâ€™s probability density. The plot is customized with orange-colored bars, black edges, axis labels, a title, and grid lines for clarity. This visualization helps to understand the skewed nature and spread of an exponential distribution

```python
#Exponential Distribution
scale = 2
size = 1000

data = np.random.exponential(scale=scale, size=size)

sns.histplot(data, bins=30, kde=True, color='orange', edgecolor='black')

plt.title(f"Exponential Distribution (Scale={scale})")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)

plt.show()
```

# Power Law Distributions
This code generates a set of random samples that follow a power-law distribution using the inverse transform sampling method. The parameters alpha, xmin, and n define the exponent, minimum value, and sample size. A uniform random variable is transformed into power-law distributed data, which is then visualized using a histogram plotted on a logarithmic scale. The resulting graph highlights the heavy-tailed nature of power-law behavior.

```python
#Power Law Distributions
# Parameters
alpha = 2.5      # power-law exponent
xmin = 1         # minimum value
n = 10000        # number of samples

# Generate power-law distributed random samples
# x = xmin * (1 - r) ** (-1 / (alpha - 1))
r = np.random.random(n)
data = xmin * (r ** (-1 / (alpha - 1)))

plt.hist(data, bins=100, density=True, alpha=0.7, log=True)
plt.xlabel("x")
plt.ylabel("P(x)")
plt.title("Power-Law Distributed Samples")
plt.show()
```

# Grouped Bar Chart
This code generates a grouped bar chart to compare two sets of values across the same categories. Bar positions are calculated using numpy to place two bars side by side for each category. Labels, a title, and a legend are included to distinguish between the two groups and clearly visualize the comparison.

```python
#Bar Chart
categories = ['Category A', 'Category B', 'Category C']
values = [15, 24, 30]

plt.barh(categories, values, color='skyblue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Basic Vertical Bar Graph')
plt.show()
```

```python
#Grouped Bar Chart
# Defining categories and their corresponding values for two groups
categories = ['Category A', 'Category B', 'Category C']
values1 = [15, 24, 30]
values2 = [20, 18, 25]

# Setting the width of the bars
bar_width = 0.35
# Calculating bar positions for both groups
bar_positions1 = np.arange(len(categories))
bar_positions2 = bar_positions1 + bar_width

# Creating the first set of bars (Group 1)
plt.bar(bar_positions1, values1, width=bar_width, label='Group 1', color='skyblue')
# Create the second set of bars (Group 2) next to the first set
plt.bar(bar_positions2, values2, width=bar_width, label='Group 2', color='orange')

# Adding labels to the axes
plt.xlabel('Categories')
plt.ylabel('Values')
# Adding a title to the graph
plt.title('Grouped Bar Graph')
# Displaying a legend to identify the groups
plt.legend()
# Showing the plot
plt.show()
```

# Stacked Bar Chart
This code produces a stacked bar chart to show the combined contribution of two groups within each category. The first set of bars (Group 1) is plotted normally, while the second set (Group 2) is stacked on top by using the bottom parameter. This visualization helps compare both individual and total values across categories.

```python
# Stacked Bar Chart
# Creating the first set of bars (Group 1) without any offset
plt.bar(categories, values1, label='Group 1', color='maroon')

# Creating the second set of bars (Group 2) plotted with 'bottom' set to the values of Group 1
# This makes Group 2 bars stacked on top of Group 1 bars
plt.bar(categories, values2, bottom=values1, label='Group 2', color='silver')

# Adding labels to the axes
plt.xlabel('Categories')
plt.ylabel('Values')
```

# Zipfâ€™s Law
This code demonstrates Zipfâ€™s Law by analyzing word frequencies in a given text. The text is first converted to lowercase and split into individual words. A dictionary is used to count how many times each word appears, and the words are then sorted in descending order of frequency. Each word is assigned a rank based on its position in the sorted list. Finally, the frequency of each word is plotted against its rank to visualize the characteristic Zipf distribution.

```python
import re
text = "Two friends were met by a bear. One climbed a tree, abandoning the other. The other played dead, and the bear left him unharmed."

#Convert text to lower case
text = text.lower()

#Remove the unwanted characters
textList = re.split(', | | . ' , text)

textDict = {}
wordFrequency={}

#Find the frequency of words
for txt in textList:
  if txt in textDict.keys():
    textDict[txt]+=1
  else:
    textDict[txt]=1

#Sort the word frequencies in descending order
wordFrequency = dict(
        sorted(
            textDict.items(),
            key=lambda x: x[1],
            reverse=True)
        )

#Define two lists, rank and frequency
rank = []
frequency = []
init = 0

#Assign ranks based on frequencies of words
for freq in wordFrequency.values():
  init+=1
  rank.append(init)
  frequency.append(freq)

#Plot the rank and frequency
plt.plot(rank,frequency)

# Labelling the x axis
plt.xlabel('Rank(r)')
# Labelling the y axis
plt.ylabel('Frequency(f)')

# Providing a title to the graph
plt.title("Zipf's law")

plt.show()
```
