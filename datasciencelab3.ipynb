```python
# Import libraries
import pandas as pd
import numpy as np
from sklearn.datasets import load_diabetes
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Define the dataset
x=np.array([1,3,5,7,8,9,10,15])
y=np.array([10,20,30,40,50,60,70,80])
```

The `person_correlation(X, Y)` function computes the **Pearson correlation coefficient** between two arrays. It measures the strength and direction of their linear relationship by dividing covariance by the product of their standard deviations. The result ranges from **-1 (perfect negative)** to **+1 (perfect positive)**, with **0** indicating no correlation.


```python
# Correlation using function
def person_correlation(X,Y):
  if len(X)==len(Y):
    Sum_xy=sum((X-X.mean())*(Y-Y.mean()))
    Sum_x_squared=sum((X-X.mean())**2)
    Sum_y_squared=sum((Y-Y.mean())**2)
    corr=Sum_xy / np.sqrt(Sum_x_squared*Sum_y_squared)
  return corr

print(person_correlation(x,y))
```

`np.corrcoef(x, y)` calculates the **Pearson correlation coefficient** using NumPy. It returns a 2×2 matrix, where the off‑diagonal values show the correlation between `x` and `y`, ranging from **-1 (negative)** to **+1 (positive)**.


```python
# Correlation using numpy
print(np.corrcoef(x,y))
```


This code plots a **heatmap of the correlation matrix** for `df1` using Seaborn.  
`vmin` and `vmax` fix the scale between **-1 and 1**, while `cmap='coolwarm'` shows negative correlations in blue and positive in red.  
`annot=True` displays the correlation values inside each cell, and `plt.show()` renders the plot.  


```python
sns.heatmap(df1.corr(), vmin=-1, vmax=1,
            annot=True, cmap='coolwarm')
plt.show()
```

```python
df=load_diabetes(as_frame=True)
df=df.frame
df.head()
```

```python
corr=df.corr(method='pearson')
corr
```

This code visualizes the correlation matrix using a Seaborn heatmap.
figsize and dpi control the plot size and resolution.
annot=True shows correlation values (formatted to 2 decimals), and linewidth=.5 adds grid lines for clarity.
plt.show() displays the heatmap.


```python
# visualize correlation using heatmap
plt.figure(figsize=(10,8), dpi=500)
sns.heatmap(corr, annot=True, fmt=".2f", linewidth=.5)
plt.show()
```

# Conclusion  
The heatmap clearly shows the **strength and direction of correlations** among features in the dataset. Positive values indicate features that increase together, while negative values highlight inverse relationships. This helps quickly identify which variables are most strongly related and useful for further analysis or modeling.
