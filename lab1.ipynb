```python

def calculate_bmi(weight, height_cm):
    # Convert height to meters
   

    # Calculate BMI
    bmi = weight / (height_cm ** 2)

    # Determine if BMI is considered healthy
    # According to the WHO, a healthy BMI is between 18.5 and 24.9
    is_healthy = 18.5 <= bmi <= 24.9

    return bmi, is_healthy
```
