```python
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
def load_dataset(file_path):
    return pd.read_csv(file_path)

# Specify the dataset path
df = load_dataset(r"C:\Users\user\Desktop\ai lab\Health_heart_experimental.csv")

# Display the first 5 rows of the dataset
df.head(5)
```

```python
# Line chart: Age vs SysBP
plt.figure(figsize=(10, 6))
sns.lineplot(data=df.sample(100), x="age", y="SysBP", marker="o")
plt.title("Line Chart: Age vs Systolic Blood Pressure")
plt.xlabel("Age")
plt.ylabel("Systolic Blood Pressure (SysBP)")
plt.grid()
plt.show()
```

baechart

```python
#barplot
sns.barplot(x=df['age'][:5], y=df['SysBP'][:5])
plt.title("Age vs Systolic Blood Pressure")
plt.xlabel("Age")
plt.ylabel("Systolic Blood Pressure (SysBP)")
plt.show()
```

```python
df['sex'].value_counts().plot(kind='pie', autopct='%0.2f', figsize=(8, 8))
plt.title("Distribution of Sex")
plt.ylabel('')
plt.show()
```

```python
sns.boxplot(df['age'])
plt.title("Boxplot of Age")
plt.show()
```

```python
sns.boxplot(x='sex', y='age', hue='indication', data=df)
plt.show()
```

```python
sns.displot(df['age'], kde=True)
plt.show()
```

```python
sns.kdeplot(df[df['indication'] == 0]['age'], label='Indication 0')
sns.kdeplot(df[df['indication'] == 1]['age'], label='Indication 1')

plt.title('Age Distribution by Indication')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
sns.scatterplot(x='age', y='SysBP', hue='sex', style='indication', data=df)
plt.title('Age vs Systolic Blood Pressure')
plt.xlabel('Age')
plt.ylabel('Systolic Blood Pressure')
plt.show()
```

```python
sns.heatmap(pd.crosstab(df['sex'], df['indication']), annot=True, cmap="twilight", fmt='d')
plt.show()
```
