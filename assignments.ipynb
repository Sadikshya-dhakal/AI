```python

# BMI Calculation Program
# Formula: BMI = weight / (height_in_meters ** 2)

def calculate_bmi(weight, height_cm):
    # Convert height to meters
    height_m = height_cm / 100

    # Calculate BMI
    bmi = weight / (height_m ** 2)

    # Determine if BMI is considered healthy
    # According to the WHO, a healthy BMI is between 18.5 and 24.9
    is_healthy = 18.5 <= bmi <= 24.9

    return bmi, is_healthy

# Input: weight in kg, height in cm
weight = float(input("Enter your weight in kilograms (kg): "))
height_cm = float(input("Enter your height in centimeters (cm): "))

# Calculate BMI and health status
bmi, is_healthy = calculate_bmi(weight, height_cm)

# Output results
print(f"Your BMI is: {bmi:.2f}")
print(f"Is your BMI considered healthy? {is_healthy}")
```

```python

# Arithmetic Operation Program

# Input: Two numbers and an operator
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
operator = input("Enter an operator (+, -, *, /): ")

# Perform the operation based on the input operator
if operator == '+':
    result = num1 + num2
elif operator == '-':
    result = num1 - num2
elif operator == '*':
    result = num1 * num2
elif operator == '/':
    if num2 != 0:
        result = num1 / num2
    else:
        result = "Error: Division by zero is not allowed."
else:
    result = "Invalid operator."

# Output the result
print(f"The result is: {result}")
```

```python

# Password Validation Program

# Input: Password
password = input("Enter your password: ")

# Check if the password meets the criteria
is_valid = (
    len(password) >= 8 and
    any(char.isalpha() for char in password) and
    any(char.isdigit() for char in password)
)

# Output the result
if is_valid:
    print("The password is valid.")
else:
    print("The password is invalid.")
```
